```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import ortho_group
import pandas as pd
from tqdm import tqdm
plt.rcParams['axes.spines.top'] = False  
plt.rcParams['axes.spines.right'] = False
plt.rcParams['grid.color'] = 'lightgrey'
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['pdf.fonttype'] = 42

figdir = '/mnt/home/awakhloo/ceph/population_geometry_opt_coding/figures/pngs'
```

```python
def theta(x): 
    return (x>0)*1

def theory_part(Psi, Phi, Omega, P):
    T1 = np.pi/P * np.trace(Psi@Psi.T) * np.trace(Omega) 
    T2 = 2 * (np.trace(Phi.T @ Psi @ Phi) - np.trace(Phi.T@Phi)**2/np.trace(Omega)) 
    T3 = np.sqrt(2) * np.trace(Phi@Phi.T)
    return 1/np.pi * np.arctan(np.sqrt(np.trace(Omega)*(T1+T2))/T3)

def emp_error(points, latents, P, n_task=300): 
    P_full, D = latents.shape 
    teach = np.random.randn(D, n_task)
    # P x n_task labels 
    labels = np.sign(latents @ teach) 
    errs = []
    idx = np.arange(P_full)
    train_idx = np.random.choice(idx, replace=False, size=(P,))
    # sample train and test set 
    test_idx = idx[~np.isin(idx, train_idx)]
    trainp, testp = points[train_idx], points[test_idx] 
    trainl, testl = labels[train_idx], labels[test_idx] 
    # n_task x N of readouts
    W = 1/P * np.einsum('ms, mk -> sk', trainl, trainp)
    yhat = np.sign(np.einsum('sk, mk -> ms', W, testp))
    errs = np.mean(theta(-testl * yhat), axis=0)
    return errs.mean(), errs.std(), errs

def draw_latents(d, alpha, p, q): 
    '''
    draw a set of latent variables with a power law covariance spectrum
    '''
    oms = q * (np.arange(d) + 1)**(-alpha) 
    samps = np.random.randn(p, d) 
    return samps * np.sqrt(oms), oms 

def draw_random_proj_x(latents, n):
    _, d = latents.shape 
    J = np.random.randn(d, n) / np.sqrt(n) 
    return latents @ J 

def draw_whitened(latents, oms, beta=1/3): 
    _, d = latents.shape 
    return latents * oms**(-beta) 

def get_covs(x, z): 
    p, _ = x.shape 
    x, z = x-x.mean(0), z - z.mean(0) 
    return x.T@x/p, x.T @ z / p, z.T @ z / p 
```

```python
np.random.seed(23534)
alphas = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5] 
ptrains = np.logspace(1,np.log10(500),7).astype(int)
d = 40 
n = 80 
p = 1000 
rep = 30
base = 5

# alphas = [0., 0.25, 0.5, 0.75, 1.0, 1.25, 1.5] 
# ptrains = np.logspace(1,np.log10(500),7).astype(int)
# d = 2 
# n = 22
# p = 1000 
# rep = 30
# base = 5

errs_emp = np.zeros((len(ptrains), len(alphas), rep, 3))
errs_the = np.zeros((len(ptrains), len(alphas),rep,3))
for q, ptrain in tqdm(enumerate(ptrains)): 
    for i, alpha in enumerate(alphas): 
        for j in range(rep): 
            z, oms = draw_latents(d, alpha, p, base)
            x_rand = draw_random_proj_x(z, n) 
            x_wh = draw_whitened(z, oms)
            errs_emp[q,i,j,0] = emp_error(x_rand, z, ptrain)[0]
            errs_emp[q,i,j,1] = emp_error(x_wh, z, ptrain)[0]
            errs_emp[q,i,j,2] = emp_error(z, z, ptrain)[0]
            Psi_r, Phi_r, Omega = get_covs(x_rand,z) 
            Psi_w, Phi_w, Omega = get_covs(x_wh, z) 
            errs_the[q,i,j,0] = theory_part(Psi_r, Phi_r, Omega, ptrain)
            errs_the[q,i,j,1] = theory_part(Psi_w, Phi_w, Omega, ptrain)
            errs_the[q,i,j,2] = theory_part(Omega, Omega, Omega, ptrain)
```

```python
ratio = 9/20 
w = 12
fig, ax = plt.subplot_mosaic([['a', 'b', 'c', 'd'], ['e', 'f', 'g', 'h']],
                             figsize=(w,ratio * w))


np.random.seed(12)
fn=14
opaq=0.2
lw=1.5
spaces = ' '
resamp = np.linspace(0.2,0.8,len(alphas))
cmap = plt.get_cmap('RdPu')(resamp)
kwargs_emp = {'ls' : '', 'marker' : '*', 'markersize' : 6.}
kwargs_the = {'lw' : lw}
X, Y = np.linspace(-1, 1, 100), np.linspace(-1, 1, 100)
X, Y = np.meshgrid(X, Y) 
levels = [0.1]
# random rotations for the contour plots 
Oz = ortho_group(2).rvs(1) 
Oxr = ortho_group(2).rvs(1) 
Oxw = ortho_group(2).rvs(1) 

def f(X,Y,O,eigs): 
    vec = np.stack([X,Y], axis=0)
    vec = np.einsum('ij, jkl -> ikl', O, vec)
    return np.linalg.norm(np.sqrt(eigs)[:, None, None] * vec, axis=0)**2

for i, alpha in enumerate(alphas): 
    # contour line for first two coords of z 
    oms = np.arange(1,d+1)**(-alpha)
    # om_eff = oms[:2]
    om_eff = np.array([oms[0], oms[np.minimum(5, d-1)]])
    # om_eff = np.array([oms[0], oms[1]]) 
    ax['b'].contour(X, Y, f(X,Y,Oz,om_eff), levels=levels, colors=[cmap[i]], linewidths=[lw], label=r'$\alpha = ${}'.format(alpha))
    ax['c'].contour(X, Y, f(X,Y,Oxr,om_eff), levels=levels, colors=[cmap[i]], linewidths=[lw])
    ax['d'].contour(X, Y, f(X,Y,Oxw,np.ones(2)), levels=[0.4], colors=[cmap[i]], linewidths=[lw])

    # plot the spectrum 
    ax['e'].loglog(np.arange(1, d+1), oms, color=cmap[i], label=r'$\alpha = ${}'.format(alpha), **kwargs_the)

    # z=x 
    ax['f'].plot(ptrains, errs_emp[:, i, :, 2].mean(1), **kwargs_emp, color=cmap[i])
    ax['f'].plot(ptrains, errs_the[:, i, :, 2].mean(1), **kwargs_the, color=cmap[i], label=r'$\alpha = ${}'.format(alpha))

    # random projection case
    ax['g'].plot(ptrains, errs_emp[:, i, :, 0].mean(1), **kwargs_emp, color=cmap[i])
    ax['g'].plot(ptrains, errs_the[:, i, :, 0].mean(1), **kwargs_the, color=cmap[i], label=r'$\alpha = ${}'.format(alpha))

    # partially whitened variables 
    ax['h'].plot(ptrains, errs_emp[:, i, :, 1].mean(1), **kwargs_emp, color=cmap[i])
    ax['h'].plot(ptrains, errs_the[:, i, :, 1].mean(1), **kwargs_the, color=cmap[i], label=r'$\alpha = ${}'.format(alpha))

    
# labels  and stuff
# ax['b'].set_title('b.)' + spaces + 'z', fontsize=fn, loc='left', pad=10)
ax['b'].set_title('z', fontsize=fn)
ax['b'].set_xlabel(r'$z_1$', fontsize=fn)
ax['b'].set_ylabel(r'$z_2$', fontsize=fn)

# ax['c'].set_title('c.)' + spaces + '$x_{rand}$', fontsize=fn, loc='left', pad=10)
ax['c'].set_title('$x^{random}$', fontsize=fn) 
ax['c'].set_xlabel(r'$x^{random}_1$', fontsize=fn)
ax['c'].set_ylabel(r'$x^{random}_2$', fontsize=fn)

# ax['d'].set_title('d.)' + spaces + '$x_{whiten}$', fontsize=fn, loc='left', pad=10)
ax['d'].set_title('$x^{whiten}$', fontsize=fn)
ax['d'].set_xlabel(r'$x^{whiten}_1$', fontsize=fn)
ax['d'].set_ylabel(r'$x^{whiten}_2$', fontsize=fn)

# config ax params 
names = ['b', 'c','d']
for n in names:
    ax[n].spines['left'].set_visible(False)
    ax[n].spines['bottom'].set_visible(False)
    # ax[n].set_xticks([])
    # ax[n].set_yticks([])
    ax[n].set_xticklabels([])
    ax[n].set_yticklabels([])
    ax[n].grid(which='both')
    ax[n].set_aspect(1)

# names_ax = ['e', 'f', 'g', 'h']
# for i in range(len(names_ax)):
#     ax[names_ax[i]].set_title(names_ax[i] + '.)', loc='left', fontsize=fn, pad=10)
# subplot_labels = iter(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
subplot_labels = iter(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

for i, a in enumerate(fig.axes):
    # Add label to upper left corner
    label = next(subplot_labels)
    a.text(-0.1, 1.2, f'{label}', transform=a.transAxes,
            fontsize=fn, va='top', ha='left', weight='bold')
    a.tick_params(labelsize=fn*3/4)


# bottom row 
ax['e'].set_ylabel(r'eigenvalues $(\Omega)$', fontsize=fn)
ax['e'].set_xlabel('eigenvalue index', fontsize=fn)

ax['f'].set_xlabel('$p$', fontsize=fn)
ax['f'].set_ylabel('$E_g$', fontsize=fn)
ax['g'].set_xlabel('$p$', fontsize=fn)
ax['g'].set_ylabel('$E_g$', fontsize=fn)
ax['h'].set_xlabel('$p$', fontsize=fn)
ax['h'].set_ylabel('$E_g$', fontsize=fn)

# set to log scales
ax['f'].set_xscale('log')
ax['g'].set_xscale('log')
ax['h'].set_xscale('log')
# make the yscales the same 
ax['f'].set_ylim(ax['h'].get_ylim())
ax['g'].set_ylim(ax['h'].get_ylim())


### schematic components 
ax['a'].axis('off')
ax['e'].legend(fontsize=fn*2/3) 
plt.tight_layout()
# fig.savefig('/Users/albertwakhloo/Documents/GitHub/abstraction/figs/gauss_skeleton.pdf', dpi=300, transparent=True)
plt.savefig(figdir + '/sm_gauss_skeleton.pdf',transparent=True)
# plt.savefig('/Users/albertwakhloo/Documents/GitHub/abstraction/figs/gauss_skeleton.svg')
```
