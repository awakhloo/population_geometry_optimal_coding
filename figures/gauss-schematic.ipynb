```python
import numpy as np 
import matplotlib.pyplot as plt 
figdir='/Users/albertwakhloo/Documents/GitHub/population-geometry-opt-coding/figures/pngs'
```

```python
fig=plt.figure()#figsize=(3,3))
mu, sig = 1, 1 
L=5
lw=2
c2, c1 = '#f4b651', '#f456de'
phi_1 = lambda x : np.exp(-(x-mu)**2/(2*sig**2))/ np.sqrt(2*np.pi*sig**2)
phi_2 = lambda x : np.exp(-(x+mu)**2/(2*sig**2))/ np.sqrt(2*np.pi*sig**2)
alpha=0.5

X = np.linspace(-L,L,1000)
plt.plot(X, phi_1(X), c=c1, lw=lw)
plt.plot(X, phi_2(X), c=c2, lw=lw)
plt.fill_between(X, 0, phi_1(X), color=c1, alpha=alpha)
plt.fill_between(X, 0, phi_2(X), color=c2, alpha=alpha)
plt.axis('off')
fig.savefig(figdir + '/wide-gauss.pdf', dpi=300, bbox_inches='tight') 
```
